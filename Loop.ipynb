```python
#Loop is a programming concept used to repeat a block of code mutiple times util a condtion is met.
#Loop help you avoid writing the same code again and again
```

```python
#For Loop: Used when you know in advance how many times you want to repeat a code
#Example

for i in range (5):
  print(i)
```

```python
sales = [100,200,300,400]
for amount in sales:
  print(amount)
```

```python
letters = ["A", "B", "C"]

for let in letters:
  print(let)
```

```python
product = ["Book", "Pen", "laptop"]

for item in product:
  print(item)
```

```python
#while loop: Used when you don't know how many times to loop, but you know the condtion to stop
```

```python
count =  1
while count <= 5:
  print(count)
  count +=1
```

```python
count =  1
while count <= 5:
  print(count)
  count +=2
```

```python
password = " "

while password != "admin123":
  password = input("Enter password:")
```

```python
#Nested loop: A loop inside another loop. commonly used for tables
```

```python
for i in range (1,4):
  for j in range(1,4):
    print(i,j)
```

```python
#break loop
for i in range(10):
  if i ==5:
    break
  print(i)
```

```python
#continue loop
for i in range(6):
  if i==2:
    continue
  print(i)
```

```python
#pass loop
for i in range (3):
  pass
```
